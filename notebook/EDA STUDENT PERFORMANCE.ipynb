```python
df["Total Score"] = df["math_score"] + df["reading_score"] + df["writing_score"]
```

```python
df["Average Score"] = df["Total Score"] / 3
```

```python
categorical_column = df.select_dtypes(include="O")
numerical_column = df.select_dtypes(exclude="O")
```

```python
fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(20,10), dpi=200)

ax1 = ax[0,0]
ax2 = ax[0,1]
ax3 = ax[1,0]
ax4 = ax[1,1]
ax5 = ax[2,0]

sns.countplot(data=df, x="gender", ax=ax1, palette="winter")
sns.countplot(data=df, x="race_ethnicity", ax=ax2, palette="winter")
sns.countplot(data=df, x="parental_level_of_education", ax=ax3, palette="winter")
ax3.tick_params(rotation=90)
sns.countplot(data=df, x="lunch", ax=ax4, palette="winter")
sns.countplot(data=df, x="test_preparation_course", ax=ax5, palette="winter")

plt.tight_layout()
```

```python
fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(20,10), dpi=200)

ax1=ax[0,0]
ax2=ax[0,1]
ax3=ax[1,0]
ax4=ax[1,1]
ax5=ax[2,0]

sns.histplot(data=df, x="Total Score", hue="gender", kde=True, ax=ax1)
ax1.set_ylabel("Score")
sns.histplot(data=df, x="Total Score", hue="race_ethnicity", kde=True, ax=ax2)
ax2.set_ylabel("Score")
sns.histplot(data=df, x="Total Score", hue="parental_level_of_education", kde=True, ax=ax3)
ax3.tick_params(rotation=90)
ax3.set_ylabel("Score")
sns.histplot(data=df, x="Total Score", hue="lunch", kde=True, ax=ax4)
ax4.set_ylabel("Score")
sns.histplot(data=df, x="Total Score", hue="test_preparation_course", kde=True, ax=ax5)
ax5.set_ylabel("Score")

plt.tight_layout()
```
